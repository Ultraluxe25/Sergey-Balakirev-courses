### Инкремент работает гораздо быстрее увеличения на единицу. То же касается и дикремента.


### Префиксный инкремент при присваивании значения дает новой переменной увеличенное на 1 значение, а постфиксальный лишь то, которое присваивали.


### Префиксный инкремент (дикремент) имеет приоритет выше умножения (деления), а постфиксный меньше

### Инкремент и дикремент нельзя применять к константам вроде чисел или выражений, а только к изменяемым вроде переменным
