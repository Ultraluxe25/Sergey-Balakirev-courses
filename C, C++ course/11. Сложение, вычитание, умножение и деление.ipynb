### В С можно складывать переменные разных типов. Все типы приводятся к int. Все вещественные приводятся к типу double. Если работаем с числами больше, то будет происходит приведение к long int (long long int) или long double


### В двух данных строчках сначала будет приведение к типам int и double, а затем понижение типов до short и float. Но чтобы не происходила потеря данных лучше использовать int и double

### Компьютер не может выполнять операции над целыми и вещественными типами сразу, а только отдельно над одинаковыми типами. Поэтому он занимается приведением типов
