# Постарайтесь пройти курс полностью. Предупреждаю, что это не простая, но вполне посильная задача, поэтому запаситесь терпением и не сдавайтесь, если что-то не получается с первого или даже с десятого раза. При этом, конечно, помните, что основная цель прохождения курса состоит в получении новых знаний и навыков, а не в формальном подтверждении в виде сертификата с надписью 100%. Сертификат с круглой цифрой — это приятный бонус, а не самоцель.
