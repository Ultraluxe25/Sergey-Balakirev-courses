#### Simple example


Templates with class

```Python
from jinja2 import Template


class Person:
    def __init__(self, title: str, years: int):
        self.title = title
        self.years = years


Sasha = Person("Master", 27)

tm = Template("I am {{ me.years - 10 }} years old and I am Kaggle {{ me.title.lower() }}")
msg = tm.render(me=Sasha)  # Use in 'tm' wit link 'me'

print(msg)


#### Same but using getters in OOP
```Python
from jinja2 import Template


class Person:
    def __init__(self, title: str, years: int):
        self.title = title
        self.years = years

    def get_title(self):
        return self.title

    def get_years(self):
        return self.years


Sasha = Person("Master", 27)

tm = Template("I am {{ me.get_years() - 10 }} years old and I am Kaggle {{ me.get_title().upper() }}")
msg = tm.render(me=Sasha)  # Use in 'tm' with link 'me'

print(msg)
```
