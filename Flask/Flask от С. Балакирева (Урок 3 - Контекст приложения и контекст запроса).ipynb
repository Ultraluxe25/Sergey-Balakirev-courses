Когда сервер получает запрос от клиента, то сначала может быть создан **контекст приложения**, а затем обязательно создается **контекст запроса**, который и является фактически представлением (обработчиком) текущего запроса.

 С контекстом приложения связаны две глобальные переменные: **g** и **current_app**. В первой сохраняется общая временная пользовательская информация, необходимая для обработки запросов. Например, мы собираемся из функций обращаться к БД. Тогда было бы правильно при появлении запроса установить соединение с БД и сохранить это соединение в какой-либо переменной, а после обработки запроса – закрыть соединение с БД.

Если фреймворк Flask видит, что мы в обработчике не собираемся использовать функционал контекста приложения, то оно не создается, чтобы не тратить процессорное время и ресурсы сервера.
