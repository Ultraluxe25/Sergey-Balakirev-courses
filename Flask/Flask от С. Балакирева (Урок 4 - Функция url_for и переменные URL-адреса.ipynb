**Контекст запроса** создается сразу после того как приходит запрос на сервер

**Рендеринг** - это процесс создания визуального представления веб-страницы или элемента пользовательского интерфейса на основе данных и шаблонов.

**url_for()** позволяет генерировать URL-адрес по имени функции-обработчика. 

**Зачем она нам нужна**, мы же и так знаем эти URL, поэтому, что нам мешает их явно указать? Если мы так будем делать, то столкнемся с множеством проблем. Во-первых, URL могут измениться, и тогда нам придется везде их менять, а если где-то забудем, то это негативно скажется на отображении страниц сайта. Или же, URL могут генерироваться динамически и тогда мы в принципе не сможем их явно прописать. Наконец, один и тот же шаблон для разных приложений, возможно, должен подставлять разный URL, например, для подключения внешних CSS или JS файлов.

Как же можно в целях отладки тестировать ее работу? Для этого фреймворк Flask позволяет искусственно создавать контекст запроса без активации веб-сервера. Это делается так:
```Python
with app.test_request_context():
    print(url_for('index'))
```
