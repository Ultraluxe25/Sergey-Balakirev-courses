## <span style="color: green">Code 1
```Python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/about')
def about():
    return "<h1>Hello, Flask</h1>"


if __name__ == '__main__':
    app.run()
```

## <span style="color: green">Code 2
```Python
from flask import Flask, render_template

app = Flask(__name__)


@app.route('/')
def index():
    return render_template('index.html')


@app.route('/about')
def about():
    return render_template('about.html')


if __name__ == '__main__':
    app.run()
```
