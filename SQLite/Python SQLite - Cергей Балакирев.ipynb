# <span style="color: green">Python SQLite #3:  SELECT  INSERT     

```python
with sq.connect('saper.db') as con:
    cur = con.cursor()
    cur.execute("""INSERT INTO users (name, sex, old, score) VALUES('', 1, 19, 1000)""")
    cur.execute("""
    INSERT INTO 
    	users 
    		(name, old, score) 
    VALUES 
    	('', 32, 200);
    """)
```

```python
with sq.connect('saper.db') as con:
    cur = con.cursor()
    cur.execute("""
    SELECT
        name,
        old,
        score
    FROM
        users;
    """)
    result = cur.fetchall()  #  
    for i in result:
        print(i)
```
